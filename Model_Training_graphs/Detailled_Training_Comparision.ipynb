## Baseline + Bahdanau

```python
import pandas as pd
df = pd.read_csv('output_baseline_10K1.txt',header=None)
df.columns = ['Step','GRU+BahdanauAttention','Loss']
df['GRU+BahdanauAttention'] = df['GRU+BahdanauAttention'].apply(lambda x: x.split()[1])
df['Loss'] = df['Loss'].apply(lambda x: x.split()[1])

df['GRU+BahdanauAttention']=pd.to_numeric(df['GRU+BahdanauAttention'])
df['Loss']=pd.to_numeric(df['Loss'])
df = df[1:]
df.head()
```

```python
print('Total Training hour base model',df['GRU+BahdanauAttention'].sum()/3600)
print('Average Training time in Bahdanau Attention',df['GRU+BahdanauAttention'].sum()/df.shape[0])
```

```python
import matplotlib.pyplot as plt
df['Loss'].plot.line(figsize=(12,5))
plt.xlabel('Steps')
plt.ylabel('Loss')
plt.title('Training details : GRU + Bahdanau Attention')
```
